# Kaggle Demo — containeer-optuna v1.0.0

Demonstrates the framework on the **customer personality** dataset from Kaggle. Requires Kaggle credentials.

**Note:** This notebook downloads data from Kaggle. If you don't have credentials, the clustering demo on iris (notebook 01) shows the same pipeline without download.

```python
import containeer_optuna as co
import numpy as np
print(f'v{co.__version__}')
```

## 1. The framework's pipeline approach

Instead of manually building a clustering pipeline like the original notebook, the framework handles it all via YAML config + OptunaRunner:

```python
cfg = co.ExperimentConfig(
    name="customer_clustering", task="clustering", dataset="customer_personality",
    model="kmeans", scaler="standard_scaler", reducer="pca",
    cv=co.CVConfig(strategy="kfold", n_splits=3),
    optimization=co.OptimizationConfig(n_trials=20, direction="maximize"),
)
cfg.optimization.storage = "sqlite:///:memory:"
cfg.optimization.load_if_exists = False
print("Config ready. Run with: co.OptunaRunner(cfg).run(n_trials=20)")
print("Requires: Kaggle credentials for customer_personality dataset")
```

## 2. Same pipeline on iris (no download needed)

Shows the exact same framework pattern on the bundled iris dataset:

```python
cfg2 = co.ExperimentConfig(
    name="iris_demo", task="clustering", dataset="iris",
    model="kmeans", scaler="standard_scaler", reducer="pca",
    cv=co.CVConfig(strategy="kfold", n_splits=3),
    optimization=co.OptimizationConfig(n_trials=15, direction="maximize"),
)
cfg2.optimization.storage = "sqlite:///:memory:"
cfg2.optimization.load_if_exists = False
runner = co.OptunaRunner(cfg2)
study = runner.run(n_trials=15, show_progress_bar=False)
print(f"Best Silhouette: {study.best_value:.4f}")
print(f"Best params: {study.best_params}")
```

## 3. Model-selection across clustering families

```python
cfg3 = co.ExperimentConfig(
    name="iris_ms", task="clustering", dataset="iris",
    model="kmeans", models=["kmeans", "agglomerative", "birch"],
    scaler="standard_scaler", reducer="pca",
    cv=co.CVConfig(strategy="kfold", n_splits=3),
    optimization=co.OptimizationConfig(n_trials=20, direction="maximize"),
)
cfg3.optimization.storage = "sqlite:///:memory:"
cfg3.optimization.load_if_exists = False
runner3 = co.OptunaRunner(cfg3)
study3 = runner3.run(n_trials=20, show_progress_bar=False)
print(f"Best Silhouette: {study3.best_value:.4f}")
best_mt = study3.best_params.get("model_type", "?")
print(f"Best clustering model: {best_mt}")
```

## Summary

The framework replaces manual pipeline construction with a declarative YAML config. The same runner handles clustering, regression, classification, and even DL/NLP.
