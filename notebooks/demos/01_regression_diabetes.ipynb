# Regression Demo — containeer-optuna v1.0.0

Demonstrates the **regression** capabilities using the bundled **diabetes** dataset.
All outputs are real and visible on GitHub.

```python
import containeer_optuna as co
import numpy as np
from collections import Counter

print(f"containeer-optuna v{co.__version__}")
print(f"Models: {len(co.MODEL_CLASSES)} | Symbols: {len(co.__all__)}")
```

```python
# Load diabetes dataset (bundled, no download)
X, y = co.get_dataset("diabetes").load()
print(f"X shape: {X.shape}")
print(f"Features: {list(X.columns)}")
print(f"y range: [{y.min():.0f}, {y.max():.0f}], mean={y.mean():.1f}")
```

## 1. Available regression models

```python
from containeer_optuna.evaluation import all_model_cards
for c in all_model_cards():
    if c.kind == "regression":
        print(f"  {c.name:25s} | {c.summary}")
```

## 2. Single-model study (Ridge, R2)

```python
cfg = co.ExperimentConfig(
    name="diabetes_ridge", task="regression", dataset="diabetes",
    model="ridge", scaler="standard_scaler", metric="r2",
    cv=co.CVConfig(strategy="shuffle_split", n_splits=5, test_size=0.2),
    optimization=co.OptimizationConfig(n_trials=30, direction="maximize"),
)
cfg.optimization.storage = "sqlite:///:memory:"
cfg.optimization.load_if_exists = False

runner = co.OptunaRunner(cfg)
study = runner.run(n_trials=30, show_progress_bar=False)

print(f"Best R2: {study.best_value:.4f}")
print(f"Best params: {study.best_params}")
```

## 3. Model-selection across families

```python
cfg3 = co.ExperimentConfig(
    name="diabetes_ms", task="regression", dataset="diabetes",
    model="ridge", models=["ridge", "random_forest", "gradient_boosting"],
    scaler="standard_scaler", metric="r2",
    cv=co.CVConfig(strategy="shuffle_split", n_splits=5, test_size=0.2),
    optimization=co.OptimizationConfig(n_trials=40, direction="maximize"),
)
cfg3.optimization.storage = "sqlite:///:memory:"
cfg3.optimization.load_if_exists = False

runner3 = co.OptunaRunner(cfg3)
study3 = runner3.run(n_trials=40, show_progress_bar=False)

print(f"Best R2: {study3.best_value:.4f}")
print(f"Best model: {study3.best_params.get('model_type')}")
print(f"Best params: {study3.best_params}")

model_types = [t.params.get("model_type") for t in study3.trials if t.params.get("model_type")]
for m, c in sorted(Counter(model_types).items()):
    print(f"  {m:25s}: {c} trials")
```

## 4. Feature-set selection

```python
cfg4 = co.ExperimentConfig(
    name="diabetes_fs", task="regression", dataset="diabetes",
    model="ridge", scaler="standard_scaler", metric="r2",
    feature_sets={
        "clinical": ["age", "sex", "bmi", "bp"],
        "serum": ["s1", "s2", "s3", "s4", "s5", "s6"],
        "full": list(X.columns),
        "bmi_focus": ["bmi", "s5", "bp", "s3"],
    },
    cv=co.CVConfig(strategy="shuffle_split", n_splits=5, test_size=0.2),
    optimization=co.OptimizationConfig(n_trials=40, direction="maximize"),
)
cfg4.optimization.storage = "sqlite:///:memory:"
cfg4.optimization.load_if_exists = False

runner4 = co.OptunaRunner(cfg4)
runner4.X = X  # keep as DataFrame for feature-set slicing
runner4.y = np.asarray(y)
study4 = runner4.run(n_trials=40, show_progress_bar=False)

print(f"Best R2: {study4.best_value:.4f}")
print(f"Best feature_set: {study4.best_params.get('feature_set')}")

fs_types = [t.params.get("feature_set") for t in study4.trials if t.params.get("feature_set")]
for f, c in sorted(Counter(fs_types).items()):
    print(f"  {f:15s}: {c} trials")
```
