# Classification Demo — containeer-optuna v1.0.0

Demonstrates classification on the bundled breast_cancer dataset.

```python
import containeer_optuna as co
import numpy as np
from collections import Counter
print(f'v{co.__version__}')
```

```python
X, y = co.get_dataset("breast_cancer").load()
print(f'X: {X.shape} | y: {y.shape}')
print(f'Classes: {np.unique(y)}')
print(f'Class balance: {np.bincount(y.astype(int))}')
```

## 1. Single-model (LogisticRegression)

```python
cfg = co.ExperimentConfig(
    name="bc_lr", task="classification", dataset="breast_cancer",
    model="logistic_regression", scaler="standard_scaler", metric="accuracy",
    cv=co.CVConfig(strategy="stratified_kfold", n_splits=5),
    optimization=co.OptimizationConfig(n_trials=30, direction="maximize"),
)
cfg.optimization.storage = "sqlite:///:memory:"
cfg.optimization.load_if_exists = False
runner = co.OptunaRunner(cfg)
study = runner.run(n_trials=30, show_progress_bar=False)
print(f"Best accuracy: {study.best_value:.4f}")
print(f"Best params: {study.best_params}")
```

## 2. Model-selection across 4 classifiers

```python
cfg2 = co.ExperimentConfig(
    name="bc_ms", task="classification", dataset="breast_cancer",
    model="logistic_regression",
    models=["logistic_regression", "knn", "svc", "decision_tree_classifier"],
    scaler="standard_scaler", metric="accuracy",
    cv=co.CVConfig(strategy="stratified_kfold", n_splits=5),
    optimization=co.OptimizationConfig(n_trials=40, direction="maximize"),
)
cfg2.optimization.storage = "sqlite:///:memory:"
cfg2.optimization.load_if_exists = False
runner2 = co.OptunaRunner(cfg2)
study2 = runner2.run(n_trials=40, show_progress_bar=False)
print(f"Best accuracy: {study2.best_value:.4f}")
best_mt = study2.best_params.get("model_type", "?")
print(f"Best model: {best_mt}")
mt = [t.params.get("model_type") for t in study2.trials if t.params.get("model_type")]
for m, c in sorted(Counter(mt).items()):
    print(f"  {m:30s}: {c} trials")
```

## 3. Save the best model

```python
pipeline = runner2.fit_best_pipeline()
print(f"Fitted: {type(pipeline).__name__}")
print(f"Steps: {[n for n, _ in pipeline.steps]}")
preds = pipeline.predict(X)
print(f"Training accuracy: {float(np.mean(preds == y)):.4f}")
```

## Summary

Auto-derived StratifiedKFold, searched across 4 model families, best pipeline reconstructed on full data.
